```python
import os
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
e2e_fields_data = pd.DataFrame({  "area": 32324,
  "customer rating": 34412,
  "family friendly": 34232,
  "food": 42514,
  "name": 51426,
  "near": 28123,
  "price": 34847,
  "Type": 28285
}, index=[0]).transpose().reset_index()
```

```python
e2e_fields_data.sort_values('frequency', ascending=True).plot(kind = 'barh', x='field', y='frequency')
plt.savefig('../MLRC-2022-Report/images/e2e-field-frequencies.png', bbox_inches='tight')
```
